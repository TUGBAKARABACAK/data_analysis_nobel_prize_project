SORU 5 :  Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

Soru bana biraz karışık geldi açıkcası."cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak" ifadesi tam netleşmedi kafamda o nedenle ben anladığım haliyle kategorileri baz alarak ona göre cinsiyet,yaş ve yılları ele aldım. Anladığım şekliyle görsele döküyorum sayın hocam, sevgiler:)

```python
# 'birth_date' sütunundaki tarih formatını düzenleyerek 'age' sütunu oluşturdum.
data['birth_date'] = pd.to_datetime(data['birth_date'], errors='coerce')
data['age'] = data['year'] - data['birth_date'].dt.year

# Ödül kategorilerini çektim datasetinden
categories = data['category'].unique()

for category in categories:
    category_data = data[data['category'] == category]

    # Cinsiyet ve Yaş Dağılımı Grafikleri (aynı figürde)
    plt.figure(figsize=(15, 6))

    # Cinsiyet Dağılımı Grafik
    plt.subplot(1, 2, 1)
    sns.countplot(x='sex', data=category_data)
    plt.title(f'Cinsiyet Dağılımı - {category}')
    plt.xlabel('Cinsiyet')
    plt.ylabel('Kişi Sayısı')

    # Yaş Dağılımı Grafik
    plt.subplot(1, 2, 2)
    sns.histplot(x='age', data=category_data, bins=20, kde=True)
    plt.title(f'Yaş Dağılımı - {category}')
    plt.xlabel('Yaş')
    plt.ylabel('Kişi Sayısı')

    plt.tight_layout() 
    plt.show()

    # Yıllara Göre Ödül Dağılımı Grafik
    plt.figure(figsize=(14, 6))
    sns.countplot(x='year', data=category_data, palette='viridis')
    plt.title(f'Yıllara Göre Ödül Dağılımı - {category}')
    plt.xlabel('Yıl')
    plt.ylabel('Kişi Sayısı')
    plt.xticks(rotation=90, fontsize=8)
    plt.show()
```

Chemistry(KİMYA) kategorisinde Nobel Ödülü kazananların çoğunluğu erkektir. Yaş aralığı genel olarak 40-80 arasında görülmekte ve yoğunluğu 50'li yaşlar oluşturmaktadır. Ödül alınan yıllara bakıldığında ise düzenli bir artış görülmemekte fakat özellikle 1985 sonrasında biraz daha fazla artış olduğu yorumu yapılabilir.

Literature(EDEBİYAT) kategorisine bakıldığında ödülü kazananların çoğunluğunu erkekler oluşturmaktadır. Yaş aralığı Chemistry ile benzemekte olup 50-80 arasındadır fakat grafiğe baktığımızda burada yoğunluğu 65-70 yaşlarının oluşturduğunu söyleyebiliriz.Ödül alınan yıllarda ise grafiğe bakıldığında genel olarak yakın değerlerde seyretmiştir yıllar içinde fakat grafiği incelediğimizde 1903,1917,1966,1974 yıllarında diğer yıllaara oranla daha fazla Literature alanında ödül kazanıldığını söylemek mümkün.

SORU 6 : 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.

Bu soruda 7.soruda olduğu gibi her kategori için ayrı bir grafik olması istenmediğinden bu şekilde bir görsel oluşturdum.

Aşağıdaki grafikte ise ülke ve kategoriyi aynı grafik üzerinde görmekteyiz.

```python
# Tarih aralığındaki verileri filtreledim.
selected_years_data = data[(data['year'] >= 1938) & (data['year'] <= 1945)]

# Çubuk Grafiği (Bar Plot)
plt.figure(figsize=(14, 6))
sns.countplot(x='birth_country', hue='category', data=selected_years_data, palette='viridis')
plt.title('Nobel Ödülü Kazananların Ülkeler ve Kategorileri (1938-1945)')
plt.xlabel('Doğum Ülkesi')
plt.ylabel('Kişi Sayısı')
plt.xticks(rotation=45)
plt.legend(title='Kategori', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

Birde Nobel Ödülü kazananların ülkelerini ve kategorilerini daha net görüp daha doğru bir yorum yapabilmek adına bu şekilde bir grafik oluşturup çıkan sonucu incelemek istedim. Burada sonuçların daha yorumlanabilir bir şekilde olduğunu söylemek mümkün.

```python
# 1938-1945 yılları arasındaki verileri seçtim.
selected_years_data = data[(data['year'] >= 1938) & (data['year'] <= 1945)]

# İki grafik için subplot oluşturdum.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# İlk grafik (Ülke Dağılımı)
sns.countplot(x='birth_country', data=selected_years_data, order=selected_years_data['birth_country'].value_counts().index, palette='viridis', ax=ax1)
ax1.set_title('Nobel Ödülü Kazananların Ülkeleri (1938-1945)')
ax1.set_xlabel('Doğum Ülkesi')
ax1.set_ylabel('Kişi Sayısı')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)

# İkinci grafik (Kategori Dağılımı)
sns.countplot(x='category', data=selected_years_data, palette='viridis', ax=ax2)
ax2.set_title("1938-1945 Yılları Arasında Nobel Ödülü Kazananların Kategorileri")
ax2.set_xlabel("Kategori")
ax2.set_ylabel("Ödül Sayısı")
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)

plt.show()
```

```python
# İlgili kategorileri ve 2000 sonrasındaki verileri seçtim
selected_categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']
selected_years_data = data[(data['category'].isin(selected_categories)) & (data['year'] >= 2000)].copy()

# 'birth_date' sütunundan yaşları çıkarttım.
selected_years_data['birth_date'] = pd.to_datetime(selected_years_data['birth_date'], errors='coerce')
selected_years_data['age'] = 2024 - selected_years_data['birth_date'].dt.year

# Her bir kategori için yan yana iki çizgi grafiği oluşturdum.
for i in range(0, len(selected_categories), 2):
    plt.figure(figsize=(18, 6))

    # İlk çizgi grafiği
    plt.subplot(1, 2, 1)
    category_data_1 = selected_years_data[selected_years_data['category'] == selected_categories[i]]
    sns.lineplot(x='birth_country', y='age', data=category_data_1)
    plt.title(f'Nobel {selected_categories[i]} Ödülü Kazananların Ülkeleri ve Yaşları (2000 Sonrası)')
    plt.xlabel('Ülke')
    plt.ylabel('Yaş')
    plt.xticks(rotation=45, ha='right')

    # İkinci çizgi grafiği (eğer varsa)
    if i + 1 < len(selected_categories):
        plt.subplot(1, 2, 2)
        category_data_2 = selected_years_data[selected_years_data['category'] == selected_categories[i + 1]]
        sns.lineplot(x='birth_country', y='age', data=category_data_2)
        plt.title(f'Nobel {selected_categories[i + 1]} Ödülü Kazananların Ülkeleri ve Yaşları (2000 Sonrası)')
        plt.xlabel('Ülke')
        plt.ylabel('Yaş')
        plt.xticks(rotation=45, ha='right')

    plt.tight_layout()  # Grafikler arasındaki boşluğu ayarlamak için
    plt.show()
```

```python
# Tarih filtresi uyguladım (2000 sonrası)
selected_years_data = data[data['year'] >= 2000]

# İlgili kategorileri seçtim.
selected_categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']
selected_data = selected_years_data[selected_years_data['category'].isin(selected_categories)]

# Her bir kategori için ayrı grafikler oluşturdum.
for category in selected_categories:
    category_data = selected_data[selected_data['category'] == category].copy()  # Copy kullanarak uyarıyı önleyin
    
    # 'birth_date' sütunundaki tarih formatını düzenleyerek 'age' sütununu oluşturdum
    category_data['birth_date'] = pd.to_datetime(category_data['birth_date'], errors='coerce')
    category_data['age'] = selected_years_data['year'] - category_data['birth_date'].dt.year

    # Ülke Dağılımı ve Yaş Dağılımı Grafikleri (aynı figürde)
    plt.figure(figsize=(20, 6))

    # Ülke Dağılımı Grafik
    plt.subplot(1, 2, 1)
    sns.countplot(x='birth_country', data=category_data, palette='viridis')
    plt.title(f'{category} Kategorisindeki Nobel Ödülü Kazananların Ülkeleri (2000 sonrası)')
    plt.xlabel('Doğum Ülkesi')
    plt.ylabel('Kişi Sayısı')
    plt.xticks(rotation=90)

    # Yaş Dağılımı Grafik
    plt.subplot(1, 2, 2)
    sns.histplot(x='age', data=category_data, bins=20, kde=True)
    plt.title(f'{category} Kategorisindeki Nobel Ödülü Kazananların Yaş Dağılımı (2000 sonrası)')
    plt.xlabel('Yaş')
    plt.ylabel('Kişi Sayısı')

    plt.tight_layout()  # Grafikler arasındaki boşluğu ayarlamak için
    plt.show()
```
